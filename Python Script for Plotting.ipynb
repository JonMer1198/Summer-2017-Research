```python
plt.plot(d[:,0],d[:,1],'p-', lw= 2,)
plt.xlabel('Time (in $10{^9}$ yr)', fontsize = 12)
plt.ylabel('BH Mass (in $M_{\odot}$)',fontsize=12)
#plt.ylabel('BH Mass (Solar Masses)', fontsize = 'large')
plt.xlim(0,4.1) # the range of x and y values
plt.ylim(100000,145000)
plt.title('Mass of BH at some Time t')
plt.ticklabel_format(style= 'sci', scilimits= (0,0))
```

```python
plt.plot(d[:,0],d[:,1],'p-', lw= 2,)
plt.xlabel('Time (Gyr)', fontsize = 'large')
plt.ylabel('BH Mass (Solar Masses)', fontsize = 'large')
plt.title('Mass of BH at some Time t')
plt.axis()
plt.ticklabel_format(style= 'sci', scilimits= (0,0))
```
